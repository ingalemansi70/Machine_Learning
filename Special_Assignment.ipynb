```python
#importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Weather Dataset.csv")
df.head()
```

```python
print("The shape of weather dataset : ",df.shape)
print("The columns in the dataset : \n",df.columns)
print("\nThe data type of columns : \n",df.dtypes)
```

```python
print("\n",df.info())
print("\n",df.describe())
```

```python
new_df = (df[df.isnull()]== True).sum()
print("Misssing(null) values in the dataset : \n")
print(new_df)
```

```python
plt.plot(df["Pressure"],df["temp_mean(c)"])
plt.title("Pressure and Temperature mean")
plt.xlabel("Pressure in hPa")
plt.ylabel("Temperature in degree Celcius")
plt.show()
```

```python
sns.histplot(df["temp_mean(c)"],bins=30,kde= True)
plt.title("Distribution of Mean Temperature")
plt.xlabel("Temp Mean in degree C")
plt.ylabel("Counts")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Wind_Bearing', y='Wind_Speed', data=df)
plt.title("Wind Speed by Wind Bearing")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Wind_Speed'])
plt.title("Boxplot of Wind Speed")
plt.show()
```

```python
sns.scatterplot(data=df,x="global_radiation",y="temp_mean(c)")
plt.title("Scatter plot of Global Radiation and Mean of Temperature")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap of Weather Dataset")
plt.show()
```
