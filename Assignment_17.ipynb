```python
#importing all required libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#q1)
df = pd.read_csv("Heart_Dataset.csv")
print("first ten rows of the data : ")
print(df.head(10))
print("\nThe shape of Heart Dataset : ",df.shape)
print("\nThe basic information of the dataset : \n")
print(df.info())
```

```python
#q2)
null_val = df.isnull()
print("Null values in the dataset : \n",null_val)
NullSum = df.isnull().sum()
print("Sum of null values in each dataset : \n",NullSum)
```

```python
#q3)
duplicate = df.duplicated()
print("The duplicate rows in the dataset are : \n")
print(duplicate)
df.drop_duplicates(inplace=True)
```

```python
#q4)
zeroChol= (df["Cholesterol"]==0).sum()
print("Count of rows with zero cholesterol is : ",zeroChol)
zeroBP =(df["RestingBP"]==0).sum()
print("Count of rows with zero Resting BP is : ",zeroChol)
```

```python
#q5)
cleanChol = (df["Cholesterol"]!=0)
countChol =cleanChol.shape
meanChol = (cleanChol.mean()).round(2)
print("mean of cholestrol is : ",meanChol)
df["Cholesterol"] =df["Cholesterol"].replace(0,meanChol)
print("Cholesterol row after replacing zero with mean of cholesterol is : ")
print(df["Cholesterol"])
cleanBP = (df["RestingBP"]!=0)
countBP =cleanBP.shape
meanBP = (cleanBP.mean()).round(2)
print("mean of RestingBP is : ",meanBP)
df["RestingBP"] =df["RestingBP"].replace(0,meanBP)
print("RestingBp row after replacing zero with mean of RestingBP is : ")
print(df["RestingBP"])
```

```python
import matplotlib.pyplot as plt

fig,axes=plt.subplots(2,2,figsize=(12,8))
fig.suptitle('Histograms of Numerical Columns',fontsize=20)
axes=axes.flatten()

def plot_histograms(column_name,pos):
    axes[pos].hist(df[column_name],bins=20,color='skyblue',edgecolor='black')
    axes[pos].set_title(column_name)
    axes[pos].set_xlabel(column_name)
    axes[pos].set_ylabel('Frequency')

plot_histograms('Age',0)
plot_histograms('RestingBP',1)
plot_histograms('Cholesterol',2)
plot_histograms('MaxHR',3)

plt.tight_layout()
plt.show()
```
