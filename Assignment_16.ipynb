```python
#importing all required libraries
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("insurance.csv")
print("first ten rows of the data : \n")
df.head(10)
```

```python
df.shape
df.info()
df.describe()
```

```python
null_data = df.isnull()

def count_null(data):
    count = data.isnull().sum() 
    print("The count of null values of ", data.name, " is : ", count)

count_null(df["age"])
count_null(df["sex"])
count_null(df["bmi"])
count_null(df["children"])
count_null(df["expenses"])
```

```python
print("The numerical columns of the dataset are : \n")
numerical_df = df[["age", "bmi", "children", "expenses"]]
print(numerical_df)
print("The categorical columns of the dataset are : \n")
categorical_df = df[["sex", "smoker", "region"]]
print(categorical_df)
```

```python

sns.histplot(data =df, x="age",kde=True)
plt.title("Histogram of age")
```

```python
sns.histplot(data =df, x="children",kde=True)
plt.title("Histogram of children")
```

```python
sns.histplot(data =df, x="expenses",kde=True)
plt.title("Histogram of expenses")

```

```python
sns.countplot(data=df,x="sex")
plt.title("Sex distribution")
```

```python
sns.countplot(data=df,x="smoker")
plt.title("Smoker distribution")
```

```python
sns.countplot(data=df,x="region")
plt.title("Region Distribution")
```

```python
corr = df[["age", "bmi", "children", "expenses"]].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

```python
print("Average Charges:", df["expenses"].mean())
print("Maximum Charges:", df["expenses"].max())
print("Minimum Charges:", df["expenses"].min())
print("\nAverage Charges by Smoker:")
print(df.groupby("smoker")["expenses"].mean())
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="smoker", y="expenses", hue="sex")
plt.title("Distribution of Charges by Smoker and Sex")
plt.show()
```

```python
# 1) Finding average age and BMI of customers
print("Average Age:", df["age"].mean())
print("Average BMI:", df["bmi"].mean())

# 2) Checking how smoking affects insurance charges
print("\nAverage Charges by Smoker Status:")
print(df.groupby("smoker")["expenses"].mean())

# 3) Finding which region has the highest number of customers
print("\nRegion with most customers:")
print(df["region"].value_counts().idxmax())

# 4) Interesting observations from visualizations
print("\nObservations:")
print("1. Smokers pay significantly higher insurance charges than non-smokers")
print("2. Expenses are highly skewed with many high-value outliers")
print("3. Age and BMI show positive correlation with expenses")
print("4. Most customers belong to the southeast region")
```
