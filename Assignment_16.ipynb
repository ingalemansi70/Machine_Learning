```python
null_data = df.isnull()
null_data
def count_null(data):
    count= null_data[data].sum()
    print("The count of null values of ",data.name," is : ",count)
count_null(df["age"])
count_null(df["sex"])
count_null(df["bmi"])
count_null(df["children"])
count_null(df["expenses"])
```
