```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Heart_Dataset.csv")
print(df.head(5))
```

FastingBS and RestingECG are the least affecting factors(columns) in the dataset so we are going to remove it from the dataset so the input requirments will be less

I decided to remove these two because they have the least correlation with the target variable on observing the correlation heatmap

```python
df = df.drop(["FastingBS","RestingECG"],axis =1)
print(df.head(3))
```

```python
X= df.drop(["HeartDisease"],axis=1 ) #input variable
y = df["HeartDisease"]  #target variable
print(X)
print(y)
```

```python
numerical = ["Age", "RestingBP", "Cholesterol", "MaxHR","Oldpeak"]
categorical =["Sex", "ChestPainType", "ExerciseAngina","ST_Slope"]
```

```python
X[numerical]=X[numerical].replace(0, np.nan)
X[numerical].fillna(X[numerical].mean())
X[categorical] = X[categorical].fillna(X[categorical].mode().iloc[0])
```

```python
cat_encoded = pd.get_dummies(X[categorical], drop_first=True).astype(int)
print("Encoded Categorical:\n", cat_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_scaled = pd.DataFrame(
    scaler.fit_transform(X[numerical]),
    columns=numerical,
    index=X.index
)
```

```python
X= pd.concat([num_scaled,cat_encoded], axis=1)
print(X)
```
