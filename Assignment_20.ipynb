```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ford_car_dataset.csv")
print("Top three columns of the dataset :")
print("\n")
print(df.head(3))
print("\nThe shape of the dataset : ",df.shape)


```

```python
y = df["price"]
X = df.drop(["price"],axis=1)
print("Independant Feature : \n")
print(X.head(2))
print("\n\nDependant Feature (Target Variable) : \n")
print(y.head(2))
```

```python
encodded_df = pd.get_dummies(data=df,columns=["model","transmission","fuelType"]).astype(int)
print("First five rows of encodded dataset is :")
print(encodded_df.head(5))
```

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
scaler = StandardScaler()
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])
print(X[numerical_cols].head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print("X_train shape : ",X_train.shape)
print("X_test shape : ",X_test.shape)
print("y_train shape : ",X_train.shape)
print("y_test shape : ",X_test.shape)
```
